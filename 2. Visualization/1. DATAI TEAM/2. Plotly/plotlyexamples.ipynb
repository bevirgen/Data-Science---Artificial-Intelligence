```python
# -*- coding: utf-8 -*-
# libraries 
from plotly.offline import init_notebook_mode, iplot, plot
init_notebook_mode(connected=True)
import plotly.figure_factory as ff
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import plotly.graph_objs as go
from plotly import tools
import plotly as py
import pandas as pd
import numpy as np
```

```python
# main data
timesData = pd.read_csv('timesData.csv')
```

```python
# Citation and Teaching Rank of Top 100 Universities
df = timesData.iloc[:100,:]
# Creating trace1
trace1 = go.Scatter(
                    x = df.world_rank,
                    y = df.citations,
                    mode = "lines",
                    name = "citations",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= df.university_name
)
# Creating trace2
trace2 = go.Scatter(
                    x = df.world_rank,
                    y = df.teaching,
                    mode = "lines+markers",
                    name = "teaching",
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text= df.university_name
)

data = [trace1, trace2]
layout = go.Layout(
    title = dict(
        text = '<b>Citation and Teaching Rank of Top 100 Universities in 2011</b>',
        x = 0.5, y = 0.9,
        font = dict(family = 'Italic',
                    size = 20,
                    color = 'Black')
),
    xaxis = dict(
        title = dict(
            text = 'World Rank',
            font = dict(family = 'Italic',
                        size = 20,
                        color = 'Black')
        ),
        ticklen = 5,
        zeroline = False
    )
)
fig = dict(data=data, layout=layout)
iplot(fig)
```

```python
# Citation Rank of Top 100 Universities in 2014,2015,2016
df2014 = timesData[timesData.year == 2014].iloc[:100,:]
df2015 = timesData[timesData.year == 2015].iloc[:100,:]
df2016 = timesData[timesData.year == 2016].iloc[:100,:]

trace1 = go.Scatter(
                    x = df2014.world_rank,
                    y = df2014.citations,
                    mode = 'markers',
                    name = '2014',
                    marker = dict(color = 'rgba(144, 112, 2, 0.8)'),
                    text = df2014.university_name
)
                    
trace2 = go.Scatter(
                    x = df2015.world_rank,
                    y = df2015.citations,
                    mode = 'markers',
                    name = '2015',
                    marker = dict(color = 'rgba(255, 112, 2, 0.8)'),
                    text = df2015.university_name
)

trace3 = go.Scatter(
                    x = df2016.world_rank,
                    y = df2016.citations,
                    mode = 'markers',
                    name = '2016',
                    marker = dict(color = 'rgba(3, 112, 2, 0.8)'),
                    text = df2016.university_name
)

data = [trace1, trace2, trace3]
layout = go.Layout(title = dict(text = '<b>Citation Rank of Top 100 Universities in 2014,2015,2016</b>',
                           x = 0.5, y = 0.9,
                           font = dict(family = 'Italic',
                                       size = 20,
                                       color = 'Black')
                          ),
              xaxis = dict(title = dict(text = 'World Rank',
                                        font = dict(family = 'Italic',
                                                   size = 20,
                                                   color = 'Black')
                                       ),
                           ticklen = 5,
                           zeroline = False
                          ),
              yaxis = dict(title = dict(text = 'Citations',
                                        font = dict(family = 'Italic',
                                                   size = 20,
                                                   color = 'Black')
                                       ),
                           ticklen = 5,
                           zeroline = False
                          ),
              )
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
# Citation and Teaching Points Best of 3 Universities in 2014
df2014 = timesData[timesData.year == 2014].iloc[:3,:]

trace1 = go.Bar(
                    x = df2014.university_name,
                    y = df2014.citations,
                    name = 'Citation',
                    marker = dict(color = 'rgba(144, 112, 2, 0.8)',
                                  line = dict(color='rgb(0,0,0)',
                                              width=1.5)
                                 ),
                    text = df2014.country)
                    
trace2 = go.Bar(
                    x = df2014.university_name,
                    y = df2014.teaching,
                    name = 'Teaching',
                    marker = dict(color = 'rgba(255, 112, 2, 0.8)',
                                  line = dict(color='rgb(0,0,0)',
                                              width=1.5)
                                 ),
                    text = df2014.country)

data = [trace1, trace2]
layout = go.Layout(title = dict(text = '<b>Citation and Teaching Points Best of 3 Universities</b>',
                           x = 0.5, y = 0.9,
                           font = dict(family = 'Italic',
                                       size = 20,
                                       color = 'Black')
                          ),
              xaxis = dict(tickangle = 0),
              barmode = 'group'
             )
fig = go.Figure(data = data, layout = layout) 
iplot(fig)
```

```python
# Top 7 Universities' Students Rate
df2016 = timesData[timesData.year == 2016].iloc[:7,:]
pie1 = df2016.num_students
pie1_list = [float(each.replace(',','.')) for each in pie1] #converted comma to dot
labels = df2016.university_name

fig = go.Figure(
    data = [
        dict(
            values = pie1_list,
            labels = labels,
            name = 'Rate',
            domain = dict(x = [1, 1], y = [1, 1]),
            hoverinfo = 'label+percent',
            hole = .3,
            type = 'pie'
        )
    ],
    layout = dict(
        title = dict(
            text = '<b> Universities Number of Students Rates</b>',
            x = 0.2, y = .94,
            font = dict(
                family = 'Italic',
                size = 20,
                color = 'Black')
        ),
        annotations = [
            dict(
                font = dict(
                    size = 20,
                    family = 'Sherif',
                    color = 'Black'
                ),
                showarrow = False,
                text = 'Number of Students',
                x = .5, y = 1.1)
        ]
    )
)
iplot(fig)
```

```python
# Top 20 Universities Teaching Score with Number of Students (Size) and International Score (Color) in 2016
df2016 = timesData[timesData.year == 2016].iloc[:20,:]
num_students_size = [float(each.replace(',','.')) for each in df2016.num_students]
international_color = [float(each) for each in df2016.international]

fig = go.Figure(
    data = [
        go.Scatter(
            x = df2016.world_rank,
            y = df2016.teaching,
            mode = 'markers',
            marker = dict(
                color = international_color,
                size = num_students_size,
                showscale = True
            ),
            text = df2016.university_name
        )
    ],
    layout = go.Layout(
        title = dict(
            text = '<b> Top 20 Universities Teaching Score with Number of Students (Size) and International Score (Color) in 2016</b>',
            x = 0.49, y = .93,
            font = dict(
                family = 'Italic',
                size = 15,
                color = 'Black')
        ),
        xaxis = dict(
            title = dict(
                text = 'World Rank',
                font = dict(family = 'Italic',
                            size = 20,
                            color = 'Black')
            ),
            ticklen = 5,
            zeroline = False
        ),
        yaxis = dict(title = dict(text = 'International Score',
                                  font = dict(family = 'Italic',
                                              size = 20,
                                              color = 'Black')
                                 ),
                     ticklen = 5,
                     zeroline = False
                    )
    )
)
iplot(fig)
```

```python
# Histogram of Students-Staff ratio in 2011 and 2012
x2011 = timesData.student_staff_ratio[timesData.year == 2011]
x2012 = timesData.student_staff_ratio[timesData.year == 2012]

trace1 = go.Histogram(
    x = x2011,
    opacity = 0.75,
    name = '2011',
    marker = dict(color = 'rgba(111, 112, 2, 0.8)')
)
trace2 = go.Histogram(
    x = x2012,
    opacity = 0.75,
    name = '2012',
    marker = dict(color = 'rgba(1, 112, 2, 0.8)')
)

data = [trace1, trace2]
layout = go.Layout(
    barmode = 'overlay',
    title = dict(
        text = '<b> Students - Staff Ratio in 2011 and 2012 </b>',
        x = 0.5, y = 0.9,
        font = dict(family = 'Italic',
                    size = 20,
                    color = 'Black')
    ),
    xaxis = dict(title = 'Students - Staff Ratio'),
    yaxis = dict(title = 'Count'),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
# In the Ranking, Which Country with the most universities
x2011 = timesData.country[timesData.year == 2011]
plt.subplots(figsize=(8,8))
wordcloud = WordCloud(
    background_color='white',
    max_font_size = 300,
    width=512,
    height=384,
    max_words=15
).generate(' '.join(x2011))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('graph.png')
plt.show()
```
